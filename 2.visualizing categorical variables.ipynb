```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('housesales.xlsx')
df.head(2)
```

## categorical distribution

boxplot 

```python
sns.boxplot(x='hometype',y='parkingtype',data=df)
```

```python
sns.boxplot(x='hometype',y='listprice',hue='parkingtype',data=df)
```

```python
## boxplot for home type and sqft with parking type
sns.boxplot(x='hometype',y='sqft',hue='parkingtype',data=df)
```

```python
sns.boxplot(y='zip',x='listprice',orient='h',data=df)
```

## violinplot
